```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets.samples_generator import make_blobs
%matplotlib inline
```

```python
np.random.seed(0)
X, labels_true = make_blobs(n_samples=500)
```

```python
init = np.random.random_integers(0,len(X)-1, size=3)
```

```python
colors = ['olivedrab', 'aqua', 'tomato']
```

```python
def plot_data(data, comment):
    plt.figure()
    plt.scatter(data[:,0], data[:,1], s=15, marker='+')
    plt.text(-5, 5.5, comment)
    plt.savefig('data_blobs.png')
    plt.show()
```

```python
def plot_init(data, init, comment, colors=colors):
    plt.figure()
    plt.scatter(data[:,0], data[:,1], s=15, marker='+')
    for idx, color in zip(init, colors):
        plt.scatter(data[idx,0], data[idx,1], s=100, c=color, marker='o', linewidths=2)
    plt.text(-5, 5.5, comment)
    plt.savefig('center_initialization.png')
    plt.show()
```

```python
def plot_clusters(data, centers, labels, comment, iteration, step, colors=colors):
    plt.figure()
    for k, color in enumerate(colors):
        members = labels == k
        plt.scatter(data[members,0], data[members,1], s=15, marker='+', c=color)
        plt.scatter(centers[k,0], centers[k,1], s=100, c=color, marker='o', linewidths=2)
    plt.text(-5, 5.5, comment)
    plt.savefig('kmeans_iter_' + str(iteration) + step + '.png')
    plt.show()
```

```python
plot_data(X, "3 Gaussians with\nunit std")
plot_init(X, init, "Random initialization\n")

init_centers = prev_centers = X[init]
for k in range(1, 7):
    kmeans = KMeans(n_clusters=3, init=init_centers, max_iter=k, n_init=1)
    kmeans.fit(X)
    plot_clusters(X, prev_centers, kmeans.labels_, "Iteration %d\nRelabel points" %k, k, 'a')
    plot_clusters(X, kmeans.cluster_centers_, kmeans.labels_, "Iteration %d\nMove centroids" %k, k, 'b')
    prev_centers = kmeans.cluster_centers_
```
