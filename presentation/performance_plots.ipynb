```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import re
%matplotlib inline
```

```python
def extract_info(lines, data):
    sections = lines.replace('  ', '').split('\n\n')
    for sect in sections:
        info = parse_section(sect)
        data = data.append(info, ignore_index=True)
    return data
```

```python
def parse_section(sect):
    info = {}
    lines = sect.split('\n')
    if lines[0] == '':
        lines = lines[1:]
    if not lines:
        return info
    info['type'] = lines[0].split()[0].strip()
    first = lines[1].split(',')
    info['n'], info['d'] =  map(int, re.findall(r'(\d+)x(\d+)', first[0])[0])
    info['k'] = map(int, re.findall(r'\d+', first[1]))[0]
    info['t'] = map(int, re.findall(r'\d+', first[2]))[0]
    info['p'] = map(int, re.findall(r'\d+', first[3]))[0]
    for l in lines[2:]:
        try:
            name, rest = l.split(':')
            info[name.strip()] = float(rest.strip('s').strip())
        except ValueError:
            continue
    return info
```

```python
data['Normalized'] = data.ix[:,'Runtime'] / data.ix[:,'Total Iterations'] * 1000
```

```python
for idx, row in data.ix[data.Normalized != data.Normalized].iterrows():
    mask1 = data.d == row.d
    mask2 = data.k == row.k
    mask3 = data.n == row.n
    mask4 = data.p == row.p
    mask5 = data.t == row.t
    nsk = data.type != "SKLEARN"
    iters = data[mask1 & mask2 & mask3 & mask4 & mask5 & nsk]['Total Iterations']
    if iters.any():
        data.ix[idx,'Normalized'] = data.ix[idx,'Runtime'] / iters.values[0] * 1000
```

```python
data['Fraction_runtime'] = data.ix[:,'Communication time'] / data.ix[:,'Runtime']
```

```python
def compare_all(size, cpus=[4,20], algos=['PYTHON', 'NUMPY', 'SEQUENTIAL', 'SKLEARN', 'MPI'], col='Normalized'):
    data_sets = []
    for alg in algos:
        m1 = data.type == alg
        m2 = np.logical_or(data.n == size, data.n == size-1) 
        if alg == 'MPI' or alg == 'SKLEARN':
            if alg == 'SKLEARN':
                m3 = data.p == 1
                time = data[m1 & m2 & m3][col].values
                steps = data[m1 & m2 & m3]['k'].values
                name = alg.lower() + ' ' + str(1)
                if len(time) > 0:
                    data_sets.append((steps, time, name))
            for c in cpus:
                m3 = data.p == c
                time = data[m1 & m2 & m3][col].values
                steps = data[m1 & m2 & m3]['k'].values
                name = alg.lower() + ' ' + str(c)
                if len(time) > 0:
                    data_sets.append((steps, time, name))
        else:
            time = data[m1 & m2][col].values
            steps = data[m1 & m2]['k'].values
            name = alg.lower()
            if len(time) > 0:
                data_sets.append((steps, time, name))
    return data_sets
```

```python
data_pairs = compare_all(10000)
```

```python
plt.figure(figsize=(8,6))
for p in data_pairs:
    plt.plot(p[0], p[1], label=p[2])
    plt.yscale('log')
plt.legend(loc=1)
plt.title('Implementation Comparison: 10,000x10 data')
plt.ylabel('Normalized Time (sec/1000 iterations)')
plt.xlabel('Number of Clusters')
plt.grid()
plt.savefig('compare10k.png')
plt.show()
```

```python
def compare_strong(size, clusters=[5,10,15,20,25,30,50], col='Runtime'):
    data_sets = []
    m1 = np.logical_or(data.n == size, data.n == size-1) 
    for c in clusters:
        m2 = data.k == c
        m3 = np.logical_or(data.type == 'SEQUENTIAL', data.type == 'MPI')
        steps = data[m1 & m2 & m3]['p'].values
        time = data[m1 & m2 & m3][col].values
        name = 'C ' + str(c) + ' clusters' 
        if len(time) > 0:
            data_sets.append((steps, time, name))
        m3 = data.type == 'SKLEARN'
        steps = data[m1 & m2 & m3]['p'].values
        time = data[m1 & m2 & m3][col].values
        name = 'sklearn ' + str(c) + ' clusters' 
    return sorted(data_sets)
```
